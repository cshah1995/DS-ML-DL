```python
df_dummy_cols = pd.get_dummies(data = df_req_cols, columns = ['venue' , 'bat_team', 'bowl_team'] , drop_first= True)
```

```python
#df_dummy_cols.drop(['level_0', 'index'], axis = 1, inplace = True)
df_dummy_cols.reset_index(drop = True)
```

```python
df_dummy_cols.head()
```

```python
X_train =df_dummy_cols.drop('total', axis = 1)[df_dummy_cols['date'].dt.year <= 2016]
X_test =df_dummy_cols.drop('total', axis = 1)[df_dummy_cols['date'].dt.year >= 2017]
```

```python
y_train = df_dummy_cols[df_dummy_cols['date'].dt.year <= 2016]['total'].values
y_test = df_dummy_cols[df_dummy_cols['date'].dt.year >= 2017]['total'].values
```

```python
print(X_train.shape)
print(X_test.shape)
print(df_dummy_cols.shape)

```

```python
X_train.drop(['date', 'year'], axis = 1, inplace = True)
X_test.drop(['date', 'year'], axis = 1, inplace = True)
```

```python
from sklearn.linear_model import LinearRegression
LRModel = LinearRegression()
LRModel.fit(X_train,y_train)
y_pred = LRModel.predict(X_test)
```

```python
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test, y_pred))
print('MSE:', metrics.mean_squared_error(y_test, y_pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
SS_Residual = sum((y_test-y_pred)**2)       
SS_Total = sum((y_test-np.mean(y_test))**2)     
r_squared = 1 - (float(SS_Residual))/SS_Total
adjusted_r_squared = 1 - (1-r_squared)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
print (r_squared, adjusted_r_squared)
```

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
```

```python
lasso=Lasso()
parameters={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40]}
lasso_regressor=GridSearchCV(lasso,parameters,scoring='neg_mean_squared_error',cv=5)

lasso_regressor.fit(X_train,y_train)
print(lasso_regressor.best_params_)
print(lasso_regressor.best_score_)
```

```python
y_predL=lasso_regressor.predict(X_test)
```

```python
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test, y_predL))
print('MSE:', metrics.mean_squared_error(y_test, y_predL))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_predL)))
```

```python
SS_Residual = sum((y_test-y_predL)**2)       
SS_Total = sum((y_test-np.mean(y_test))**2)     
r_squared = 1 - (float(SS_Residual))/SS_Total
adjusted_r_squared = 1 - (1-r_squared)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
print (r_squared, adjusted_r_squared)
```
